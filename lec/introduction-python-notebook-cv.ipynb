```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading an image

```python
img = cv2.imread('lena.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))
plt.imshow(img_rgb, cmap='gray')
```

```python
print 'image shape', img.shape, 'dim', img.ndim
```

# Subimage

```python
sub_img = img_rgb[1:100,1:20,1]

plt.figure(figsize=(12,12))
plt.imshow(sub_img, cmap='gray', interpolation='none')
```

# Converting to Grayscale Image

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(5,5))
plt.imshow(img_gray, cmap='gray')
```

```python
print 'image shape', img_gray.shape, 'dim', img_gray.ndim
```

# Picking a row from the image

```python
row = img_gray[:,200]
print row.shape, row.ndim
plt.plot(row, 'b--')
```

# Creating a 2D image

```python
import numpy as np

my_img = np.array([[0,1],[1,0]], dtype='float32')

plt.figure(figsize=(5,5))
plt.imshow(my_img, cmap='gray', interpolation='none')
```

```python
my_img = np.array([[-1,1],[1,-1]], dtype='float32')

plt.figure(figsize=(5,5))
plt.imshow(my_img, cmap='Reds', interpolation='none')
```
